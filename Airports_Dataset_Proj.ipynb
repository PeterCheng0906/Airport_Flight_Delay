```python
import json
import pandas as pd
```

```python
airports_df= pd.read_csv('airports.csv')
airports_df = (
    airports_df
    .drop(columns=['home_link', 'wikipedia_link', 'keywords', 'continent', 'icao_code', 'gps_code', 'local_code']) # Drops irrelevant fields
    .dropna(subset=['iata_code']) # Removes any ports without an IATA Code
)

airports_df = airports_df[airports_df['iata_code'].str.strip() != '']
airports_df = airports_df[airports_df['type'].str.contains('airport', case=False, na=False)] # Only consider airports, not helicopter ports or balloon ports
airports_df = airports_df.rename(columns={'name': 'airport'}) 

airports_df.head()
```

```python
countries = pd.read_csv('countries.csv')
countries = countries.drop(columns = ['id', 'wikipedia_link', 'keywords', 'continent']) 
countries.head()
```

```python
airports_df = airports_df.merge(
    countries[['code', 'name']],
    left_on='iso_country',
    right_on='code',
    how='left'
)

airports_df = airports_df.rename(columns={'name': 'country'}) 
airports_df = airports_df.drop(columns=['code'])

airports_df.to_csv('data/airports_cleaned.csv', index=False)

airports_df.head()
```
