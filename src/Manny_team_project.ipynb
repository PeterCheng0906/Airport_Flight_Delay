```python
import pandas as pd
# read csv
weather = pd.read_csv("../Input/open_meteo.csv")
# dimension check
print("weather dimension:", weather.shape)
# inspecting columns and nulls
print("\nweather columns:")
print(weather.info())
print("\nweather null counts:")
print(weather.isnull().sum())
weather.head(10)
```

```python
num_unique_iata = weather["IATA_CODE"].nunique()
print(f"Unique IATA codes: {num_unique_iata}")
```

```python
weather["airport_time"] = (
    weather["IATA_CODE"] + "_" + weather["time"].astype(str)
)
weather.head(10)
```

```python
weather = weather.drop(columns=["IATA_CODE"])
weather = weather.rename(columns={
    "weather_code (wmo code)": "weather_code",
    "temperature_2m_mean (°F)": "temperature_2m_mean",
    "cloud_cover_mean (%)": "cloud_cover_mean",
    "wind_speed_10m_mean (mp/h)": "wind_speed_10m_mean",
    "relative_humidity_2m_mean (%)": "relative_humidity_2m_mean",
    "precipitation_sum (inch)": "precipitation_sum"
})


weather.to_csv(
    "../Output/manny_weather_cleaned.csv",
    index=False
)

print(weather.info())
```
